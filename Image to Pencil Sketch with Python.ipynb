## Task 4: Image to Pencil Sketch with Python

## Importing the relevant libraries

```python
import cv2
from PIL import Image
from IPython.display import display
from skimage import io
%matplotlib inline
```

```python
image = cv2.imread('Mohamad Salah.jpg')
image = cv2.resize(image,(800,800))
display(Image.fromarray(image))
```

## Converting the image into gray scale

```python
gray_image = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
display(Image.fromarray(gray_image))
```

## Inverting the image

```python
inverted_image = cv2.bitwise_not(gray_image)
display(Image.fromarray(inverted_image))
```

## Blurring the image using Gaussian Blur

```python
blurred_image = cv2.GaussianBlur(inverted_image, (5, 5), 0, 0)
display(Image.fromarray(blurred_image))
```

## Inverting the image to get the final pencil sketch

```python
inverted_blurred_image = 255 - blurred_image
pencil_sketch = cv2.divide(gray_image, inverted_blurred_image, scale=250.0)
display(Image.fromarray(pencil_sketch))
```
